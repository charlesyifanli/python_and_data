```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Using only pandas (and matplotlib/seaborn if necessary) plot the distribution for
- Age
- Sleep Duration
- Quality of Sleep
- Physical Activity Level
- Stress Level
- Heart Rate

```python
df = pd.read_csv('./sleep_health.csv')
df.columns
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

plt.sca(axes[0, 0])
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.sca(axes[0, 1])
plt.hist(df['Sleep Duration'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Sleep Duration')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Frequency')

plt.sca(axes[0, 2])
plt.hist(df['Quality of Sleep'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Quality of Sleep')
plt.xlabel('Quality of Sleep')
plt.ylabel('Frequency')

plt.sca(axes[1, 0])
plt.hist(df['Physical Activity Level'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Physical Activity Level')
plt.xlabel('Physical Activity Level')
plt.ylabel('Frequency')

plt.sca(axes[1, 1])
plt.hist(df['Stress Level'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Frequency')

plt.sca(axes[1, 2])
plt.hist(df['Heart Rate'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Heart Rate')
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
fig2, axes = plt.subplots(2, 3, figsize=(18, 10))

plt.sca(axes[0, 0])
sns.boxplot(x='Quality of Sleep', y='Sleep Duration', data=df)
plt.title('Distribution of Sleep Duration based on Quality of Sleep')
plt.xlabel('Quality of Sleep')
plt.ylabel('Sleep Duration (hours)')

plt.sca(axes[0, 1])
sns.boxplot(x='Stress Level', y='Sleep Duration', data=df)
plt.title('Distribution of Sleep Duration based on Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Sleep Duration (hours)')

plt.sca(axes[0, 2])
sns.boxplot(x='Physical Activity Level', y='Sleep Duration', data=df)
plt.title('Distribution of Sleep Duration based on Physical Activity Level')
plt.xlabel('Physical Activity Level')
plt.ylabel('Sleep Duration (hours)')

plt.sca(axes[1, 0])
sns.boxplot(x='Occupation', y='Sleep Duration', data=df)
plt.title('Distribution of Sleep Duration based on Occupation')
plt.xlabel('Occupation')
plt.ylabel('Sleep Duration (hours)')

plt.sca(axes[1, 1])
sns.boxplot(x='BMI Category', y='Sleep Duration', data=df, order=['Normal', 'Normal Weight', 'Overweight', 'Obese'])
plt.title('Distribution of Sleep Duration based on BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Sleep Duration (hours)')

plt.tight_layout()
plt.show()
```

Using only numpy, use the appropriate plot to
- See the distribution of 'Sleep Duration' based on 'Quality of Sleep'
- See the distribution of 'Sleep Duration' based on 'Stress Level'
- See the distribution of 'Sleep Duration' based on 'Physical Activity Level'
- See the distribution of 'Sleep Duration' based on 'Occupation'
- See the distribution of 'Sleep Duration' based on 'BMI'

```python
fig3, axes = plt.subplots(2, 3, figsize=(18, 10))
sleep_duration = nd['Sleep_Duration']

#1
quality_of_sleep = nd['Quality_of_Sleep']
unique_quality = np.unique(quality_of_sleep)
sleep_duration_by_quality = [sleep_duration[quality_of_sleep == q] for q in unique_quality]
##
plt.sca(axes[0, 0])
plt.boxplot(sleep_duration_by_quality, tick_labels=unique_quality)
plt.title('Distribution of Sleep Duration based on Quality of Sleep')
plt.xlabel('Quality of Sleep')
plt.ylabel('Sleep Duration (hours)')

#2
stress_level = nd['Stress_Level']
unique_stress = np.unique(stress_level)
sleep_duration_by_stress = [sleep_duration[stress_level == s] for s in unique_stress]
##
plt.sca(axes[0, 1])
plt.boxplot(sleep_duration_by_stress, tick_labels=unique_stress)
plt.title('Distribution of Sleep Duration based on Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Sleep Duration (hours)')

#3
physical_activity_level = nd['Physical_Activity_Level']
unique_activity = np.unique(physical_activity_level)
sleep_duration_by_activity = [sleep_duration[physical_activity_level == a] for a in unique_activity]
##
plt.sca(axes[0, 2])
plt.boxplot(sleep_duration_by_activity, tick_labels=unique_activity)
plt.title('Distribution of Sleep Duration based on Physical Activity Level')
plt.xlabel('Physical Activity Level')
plt.ylabel('Sleep Duration (hours)')

#4
occupation = nd['Occupation']
unique_occupation = np.unique(occupation)
sleep_duration_by_occupation = [sleep_duration[occupation == o] for o in unique_occupation]
##
plt.sca(axes[1, 0])
plt.boxplot(sleep_duration_by_occupation, tick_labels=unique_occupation)
plt.title('Distribution of Sleep Duration based on Occupation')
plt.xlabel('Occupation')
plt.ylabel('Sleep Duration (hours)')

#5
bmi = nd['BMI_Category']
unique_bmi = np.unique(bmi)
sleep_duration_by_bmi = [sleep_duration[bmi == b] for b in unique_bmi]
##
plt.sca(axes[1, 1])
plt.boxplot(sleep_duration_by_bmi, tick_labels=unique_bmi)
plt.title('Distribution of Sleep Duration based on BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Sleep Duration (hours)')

plt.tight_layout()
plt.show()
```
