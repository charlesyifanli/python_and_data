```python
import random
import time

# Defining the card deck
card_deck = {
    'Diablo': {'Health': 100, 'Attack': 90, 'Defense': 60},
    'Medusa': {'Health': 100, 'Attack': 70, 'Defense': 70},
    'Jester': {'Health': 120, 'Attack': 60, 'Defense': 90},
    'Troll': {'Health': 150, 'Attack': 40, 'Defense': 94},
    'Specter': {'Health': 100, 'Attack': 70, 'Defense': 70},
    'Mist': {'Health': 100, 'Attack': 75, 'Defense': 65},
    'Savage': {'Health': 100, 'Attack': 90, 'Defense': 50},
    'Marauder': {'Health': 100, 'Attack': 85, 'Defense': 50},
    'Wimp': {'Health': 110, 'Attack': 40, 'Defense': 85},
    'Sorcerer': {'Health': 100, 'Attack': 70, 'Defense': 55}
}

# Deck size for both players
deck_size = 2

# Draw random cards for both player and opponent
player_deck = random.sample(list(card_deck.items()), deck_size)
opponent_deck = random.sample(list(card_deck.items()), deck_size)


# Function to show the player's deck
def display_player_deck():
    print("\nYour cards:")
    for i, card in enumerate(player_deck):
        print(
            f"{i + 1}. {card[0]} - Health: {card[1]['Health']}, Attack: {card[1]['Attack']}, Defense: {card[1]['Defense']}")


# Function to process an attack
def process_attack(attacker, defender, defender_name):
    if attacker['Attack'] > defender['Defense']:
        damage_to_health = attacker['Attack'] - defender['Defense']
        defender['Defense'] = 0
        defender['Health'] -= damage_to_health
        print(f"The attack breaks through {defender_name}'s defense and deals {damage_to_health} damage to health!")
    else:
        defender['Defense'] -= attacker['Attack']
        print(f"The attack decreases {defender_name}'s defense by {attacker['Attack']}!")

    if defender['Health'] <= 0:
        print(f"{defender_name} has been defeated!")


# Game loop
while player_deck and opponent_deck:
    # Show player's deck
    display_player_deck()

    # Draw a random card from the opponent's deck
    opponent_card = random.choice(opponent_deck)
    opponent_name = opponent_card[0]
    opponent_stats = opponent_card[1]

    print(
        f"\nOpponent draws {opponent_name} - Health: {opponent_stats['Health']}, Attack: {opponent_stats['Attack']}, Defense: {opponent_stats['Defense']}")

    # Player picks a card to fight
    choice = int(input("Choose a card from your deck to fight (1-5): ")) - 1
    player_card = player_deck[choice]
    player_name = player_card[0]
    player_stats = player_card[1]

    print(f"\nYou chose {player_name} to fight!")

    # Check if player's attack defeats opponent's card in one hit
    if player_stats['Attack'] > opponent_stats['Health']:
        print(f"{player_name} defeats {opponent_name} in one hit!")
        opponent_deck.remove(opponent_card)
    else:
        # Player's turn to attack
        print(f"{player_name} attacks {opponent_name}!")
        process_attack(player_stats, opponent_stats, opponent_name)

        # Remove opponent's card if health drops to 0 or below
        if opponent_stats['Health'] <= 0:
            print(f"{opponent_name} has been removed from the opponent's deck!")
            opponent_deck.remove(opponent_card)
        else:
            time.sleep(3)
            # Opponent's turn to attack
            print(f"\nNow, {opponent_name} attacks {player_name}!")
            process_attack(opponent_stats, player_stats, player_name)

            if player_stats['Health'] <= 0:
                print(f"{player_name} has been defeated and removed from your deck!")
                player_deck.remove(player_card)

    # Pause before next round
    time.sleep(3)
    print("\nA new round will begin...")

# End of game
if not opponent_deck:
    print("Congratulations! You have won the game!")
else:
    print("The opponent has won the game!")
```
