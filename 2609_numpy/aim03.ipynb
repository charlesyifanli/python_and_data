```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Using only pandas (and matplotlib/seaborn if necessary) plot the distribution for
- Age
- Sleep Duration
- Quality of Sleep
- Physical Activity Level
- Stress Level
- Heart Rate

```python
df = pd.read_csv('./sleep_health.csv')
df.columns
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

sns.histplot(df['Age'], ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Age')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Frequency')

sns.histplot(df['Sleep Duration'], ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Sleep Duration')
axes[0, 1].set_xlabel('Sleep Duration (hours)')
axes[0, 1].set_ylabel('Frequency')

sns.histplot(df['Quality of Sleep'], ax=axes[0, 2])
axes[0, 2].set_title('Distribution of Quality of Sleep')
axes[0, 2].set_xlabel('Quality of Sleep')
axes[0, 2].set_ylabel('Frequency')

sns.histplot(df['Physical Activity Level'], ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Physical Activity Level')
axes[1, 0].set_xlabel('Physical Activity Level')
axes[1, 0].set_ylabel('Frequency')

sns.histplot(df['Stress Level'], ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Stress Level')
axes[1, 1].set_xlabel('Stress Level')
axes[1, 1].set_ylabel('Frequency')

sns.histplot(df['Heart Rate'], ax=axes[1, 2])
axes[1, 2].set_title('Distribution of Heart Rate')
axes[1, 2].set_xlabel('Heart Rate')
axes[1, 2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
nd = np.genfromtxt('./sleep_health.csv', delimiter=',', skip_header=0, dtype=None, encoding='utf-8', names=True)
nd.dtype.names
```

```python
fig2, axes = plt.subplots(2, 3, figsize=(18, 10))

sns.histplot(nd['Age'], ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Age')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Frequency')

sns.histplot(nd['Sleep_Duration'], ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Sleep Duration')
axes[0, 1].set_xlabel('Sleep Duration (hours)')
axes[0, 1].set_ylabel('Frequency')

sns.histplot(nd['Quality_of_Sleep'], ax=axes[0, 2])
axes[0, 2].set_title('Distribution of Quality of Sleep')
axes[0, 2].set_xlabel('Quality of Sleep')
axes[0, 2].set_ylabel('Frequency')

sns.histplot(nd['Physical_Activity_Level'], ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Physical Activity Level')
axes[1, 0].set_xlabel('Physical Activity Level')
axes[1, 0].set_ylabel('Frequency')

sns.histplot(nd['Stress_Level'], ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Stress Level')
axes[1, 1].set_xlabel('Stress Level')
axes[1, 1].set_ylabel('Frequency')

sns.histplot(nd['Heart_Rate'], ax=axes[1, 2])
axes[1, 2].set_title('Distribution of Heart Rate')
axes[1, 2].set_xlabel('Heart Rate')
axes[1, 2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
