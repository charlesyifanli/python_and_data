```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./sleep_health.csv')
df.columns
```

```python
nd = np.genfromtxt('./sleep_health.csv', dtype=None, delimiter=',', names=True, encoding='utf-8')
nd.dtype.names
```

the relation between
- gender and stress level

```python
plt.figure(figsize=(10, 4))

sns.boxplot(x=df['Gender'], y=df['Stress Level'])
plt.title('Gender and Stress')
plt.xlabel('Gender')
plt.ylabel('Stress')

plt.show()
```

```python
plt.figure(figsize=(10, 4))

sns.boxplot(x=nd['Gender'], y=nd['Stress_Level'])
plt.title('Gender and Stress')
plt.xlabel('Gender')
plt.ylabel('Stress')

plt.show()
```
